```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import numpy as np
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
fundname_list = []
risk_list = []
category_list = []
rating_list = []
y1_list = []
y3_list = []
y5_list = []
pagenum = []

session = requests.Session()
total_time = time.time()

for i in range(0, 111):
    start_time = time.time()

    URL = f"https://groww.in/mutual-funds/filter?q=&fundSize=&pageNo={i}"
    page = session.get(URL)
    soup = BeautifulSoup(page.text, "html.parser")

    # Each fund card
    cards = soup.find_all("tr", class_="f22Card")

    for card in cards:
        # Fund name
        name = card.find('div', class_='contentPrimary f22LH34 f22Mb4 truncate bodyBaseHeavy')
        fundname_list.append(name.text.strip() if name else np.nan)

        # Risk + Category
        rc = card.find_all('div', class_='contentSecondary f22Ls2 contentTertiary bodySmallHeavy')
        risk_list.append(rc[0].text.strip() if len(rc) > 0 else np.nan)
        category_list.append(rc[1].text.strip() if len(rc) > 1 else np.nan)

        # Rating
        rating = card.find('div', class_='contentSecondary f22Ls2 contentTertiary valign-wrapper bodySmallHeavy')
        rating_list.append(rating.text.strip() if rating else np.nan)

        # Returns
        returns = card.find_all('div', class_='contentPrimary center-align f22Mb4 bodyBaseHeavy')
        y1_list.append(returns[0].text.strip() if len(returns) > 0 else np.nan)
        y3_list.append(returns[1].text.strip() if len(returns) > 1 else np.nan)
        y5_list.append(returns[2].text.strip() if len(returns) > 2 else np.nan)

        pagenum.append(i)

    print(f'Page {i} completed in {time.time()-start_time:.2f} seconds')

print("Total Time Completed in seconds", time.time()-total_time)

df = pd.DataFrame({
    "fund_name": fundname_list,
    "risk": risk_list,
    "category": category_list,
    "rating": rating_list,
    "1y_return": y1_list,
    "3y_return": y3_list,
    "5y_return": y5_list,
})
```
